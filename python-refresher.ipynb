```python
# functions, if statements
def fibonacci(n=5):
    if n==0:
        return 1
    else:
        return n*fibonacci(n-1)
    
fibonacci()
```

# Jupyter workflow tips

* run a cell with Shift-Enter
* Jupyter displays value of last expression in cell 
* open a new cell below the current one (Alt-Enter) to see values of variables, test assumptions
* you can reorder cells using the arrows in the toolbar, or by copy-pasting cells up or down

How to troubleshoot and debug in Jupyter
* find the line in the cell that's not working as expected 
* test that every input to that line is what you expect by opening a new cell below the one that's not working, copy pasting code if needed
  * check the type (eg, integer? string? array? matrix?)
  * check the values (eg, are there NaNs or Infs? are there negative numbers where you expect positive?)
