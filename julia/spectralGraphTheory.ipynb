# Create Laplacian of Path graph

```julia
n = 15
L = convert(Array{Float64}, 2*Diagonal(ones(n)))
```

```julia
neg = [zeros(n-1) convert(Array{Float64}, -1*Diagonal(ones(n-1)))]
negUpper = vcat(neg, zeros(15)')
neg = [convert(Array{Float64}, -1*Diagonal(ones(n-1))) zeros(n-1)]
negLower = vcat(zeros(n)', neg)
```

```julia
L = L+negUpper+negLower
L[1,1]=1.
L[n,n]=1.
L
```

```julia
D, V = eig(L)
```
